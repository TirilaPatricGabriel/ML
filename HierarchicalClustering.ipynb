```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='blue', label='CL.1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='red', label='CL.2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='yellow', label='CL.3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='green', label='CL.4')
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], s=100, c='pink', label='CL.5')
plt.title('Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
